# CRC16 XMODEM

**Create activity or flowchart diagram from which we can create the program**

**Tento diagram popisuje následující kroky:**
    
- Inicializace CRC: Nastavení počáteční hodnoty CRC na 0x0000.
- Procházení dat: Procházení dat bajt po bajtu.
-Výpočet CRC: Pro každý bajt:
-        XORování aktuálního bajtu s nejvyšším bajtem CRC.
-        Procházení každého bitu bajtu.
-        Pokud je nejvyšší bit CRC nastaven na 1, XORování s polynomem 0x1021 a posun CRC o jeden bit doleva.
-        Pokud je nejvyšší bit CRC nastaven na 0, pouze posun CRC o jeden bit doleva.
-        Zajištění, že CRC zůstane 16-bitové číslo.
    
**Výsledek:** Výsledná hodnota CRC-16 po zpracování všech bajtů.
